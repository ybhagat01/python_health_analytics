```python
import numpy as np
from collections import Counter
name = []
age= []
gender = []
blood_pressure = []
temperature = []

with open("code11.txt","r") as file:
    lines = file.readlines()
    
for line in lines[1:]:
    clean_line = line.strip()
    n, a, g, bp, temp = [x.strip() for x in clean_line.split(",")]
    name.append(n)
    age.append(int(a))
    gender.append(g)
    sys, dia = bp.split("/")
    blood_pressure.append((int(sys), int(dia)))
    temperature.append(float(temp))
    
avg_age = np.mean(age)
print("Average age:",avg_age)

counts = Counter(gender)
m= counts["Male"]
f= counts["Female"]
print("No. of male patients:",m)
print("No. of female patients:",f)

hbp = max(blood_pressure)
print("Highest Blood Pressure:", str(hbp[0])+ "/"+str(hbp[1]))

lbp = min(blood_pressure)
print("Lowest Blood Pressure:", str(lbp[0])+"/"+str(lbp[1]))

avg_temp = np.mean(temperature)
print("Average temperature:", avg_temp)
```
