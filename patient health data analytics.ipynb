```python
class patient:
    def __init__(self, name, age, gender, blood_pressure, temperature):
        self.name = name
        self.age = age
        self.gender = gender
        self.blood_pressure = blood_pressure
        self.temperature = temperature
        
    def display_patient_info(self):
        print(f"\n Patient information: ")
        print (f"Patient's name: {self.name}")
        print(f"Age:{self.age}")
        print (f"Gender: {self.gender}")
        print(f"Blood pressure:{self.blood_pressure}")
        print(f"Temperature:{self.temperature}")
        
    def is_hypersensitive(self):
        systolic, diastolic = self.blood_pressure
        if systolic >=140 or diastolic >=90:
            return True
        else:
            return False
            
    def is_feverish(self):
        if self.temperature > 37.5:
            return True
        else:
            return False
        
p1 = patient("John Doe", 45, "Male", (140,85), 36.8)
p2 = patient("Jane Smith", 32, "Female", (130,88), 37.6)

p1.display_patient_info()
p2.display_patient_info()

print(f"\n{p1.name} is hypersensitive: {p1.is_hypersensitive()}")
print(f"{p1.name} is feverish: {p1.is_feverish()}")

print(f"\n{p2.name} is hypersensitive: {p2.is_hypersensitive()}")
print(f"{p2.name} is feverish: {p2.is_feverish()}")
```
