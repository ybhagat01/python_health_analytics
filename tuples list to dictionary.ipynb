```python
data_list = [("apple",5), ("banana",2), ("orange",8),("grapes",3),("pineapple",1)]

def convert_to_dictionary(data_list):
    data_dic = {}
    for fruits in data_list:
        name , quantity = fruits
        data_dic[name] = quantity
    return data_dic
data_dic = convert_to_dictionary(data_list)
print ("Orignal List of Tuples: ", data_list)
print ("Converted Dictionary: ", data_dic)
```
